# Jupyter Notebook Shortcuts

| Syntax               | Description      |
|:--------------------:|:-----------------|
| **Ctrl + ,**         | `Settings`       |
| **Ctrl + Shift + P** | `Command Palette`|

* <b> Ctrl </b>:                        Command key ⌘
* <b> Shift </b>:                       Shift ⇧
* <b> Alt </b>:                         Option ⌥

---
## There are 2 modes in the Jupyter Notebook App:
* Command Mode 
* Edit Mode

## Shortcuts in both modes:
* <b> Shift + Enter </b>:               run the current cell, select below
* <b> Ctrl + Enter </b>:                run selected cells
* <b> Alt + Enter </b>:                 run the current cell, insert below
* <b> Ctrl + S </b>:                    save and checkpoint

## Command Mode (press ESC to activate):
* <b> Enter </b>:                       take you into edit mode
* <b> H </b>:                           show all shortcuts
* <b> Up </b>:                          select cell above (UP)
* <b> K </b>:                           select cell above (UP)
* <b> Down </b>:                        select cell below (DOWN)
* <b> J </b>:                           select cell below (DOWN)
* <b> Shift + Up </b>:                  extend selected cells above
* <b> Shift + Down </b>:                extend selected cells below
* <b> A </b>:                           insert cell above
* <b> B </b>:                           insert cell below
* <b> X </b>:                           cut selected cells
* <b> C </b>:                           copy selected cells
* <b> V </b>:                           paste cells below
* <b> Shift + V </b>:                   paste cells above
* <b> D, D </b>: (press the key twice)  delete selected cells
* <b> Z </b>:                           undo cell deletion
* <b> S </b>:                           Save and Checkpoint
* <b> Y </b>:                           change the cell type to Code
* <b> M </b>:                           change the cell type to Markdown
* <b> P </b>:                           open the command palette
* <b> Shift + Space </b>:               scroll notebook up
* <b> Space </b>:                       scroll notebook down

## Edit mode (press ENTER to activate)
* <b> Esc </b>:                         take you into command mode
* <b> Tab </b>:                         code completion or indent
* <b> Shift + Tab </b>:                 tooltip
* <b> Ctrl + ] </b>:                    indent
* <b> Ctrl + [ </b>:                    dedent
* <b> Ctrl + A </b>:                    select all
* <b> Ctrl + Z </b>:                    undo
* <b> Ctrl + Shift + Z </b>:            redo
* <b> Ctrl + Y </b>:                    redo
* <b> Ctrl + Home </b>:                 go to cell start
* <b> Ctrl + End </b>:                  go to cell end
* <b> Ctrl + Left </b>:                 go one word left
* <b> Ctrl + Right </b>:                go one word right
* <b> Ctrl + Shift + P </b>:            open the command palette
* <b> Down </b>:                        move cursor down
* <b> Up </b>:                          move cursor up
